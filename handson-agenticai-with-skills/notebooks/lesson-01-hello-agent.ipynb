## Cell 1 — Import and create the client

````python
import anthropic

# Reads ANTHROPIC_API_KEY from your environment automatically
client = anthropic.Anthropic()

print("Client ready!")
````

## Cell 2 — Your first API call

We use `claude-haiku-4-5-20251001` — fastest and cheapest, perfect for learning.

````python
response = client.messages.create(
    model="claude-haiku-4-5-20251001",
    max_tokens=1024,
    messages=[
        {"role": "user", "content": "What is an AI agent? Explain in 2 sentences."}
    ]
)

print(response.content[0].text)
````

## Cell 3 — Inspect the full response object

The response has more than just text. `stop_reason` is especially important — you'll use it in the agent loop.

````python
print("Stop reason: ", response.stop_reason)       # "end_turn" = finished naturally
print("Model:       ", response.model)
print("Input tokens:", response.usage.input_tokens)
print("Output tokens:",response.usage.output_tokens)
````

## Cell 4 — Add a system prompt

The `system` parameter sets the agent's persona. It is not part of the message history.

> ☕ **Java analogy:** Like a constructor that sets the initial state of your agent object.

````python
response = client.messages.create(
    model="claude-haiku-4-5-20251001",
    max_tokens=512,
    system="You are a concise assistant who only answers in bullet points.",
    messages=[
        {"role": "user", "content": "What are the 3 parts of an AI agent?"}
    ]
)

print(response.content[0].text)
````

## Cell 5 — Multi-turn conversation

Build a conversation by appending messages. You manage the history — the API is stateless.

````python
# Start with the first turn
messages = [
    {"role": "user", "content": "My name is Tianhong. I'm learning agentic AI."}
]

resp1 = client.messages.create(
    model="claude-haiku-4-5-20251001",
    max_tokens=256,
    messages=messages
)

print("Turn 1:", resp1.content[0].text)

# Append assistant reply + new user message
messages.append({"role": "assistant", "content": resp1.content[0].text})
messages.append({"role": "user", "content": "What's my name and what am I learning?"})

resp2 = client.messages.create(
    model="claude-haiku-4-5-20251001",
    max_tokens=256,
    messages=messages
)

print("\nTurn 2:", resp2.content[0].text)
````
